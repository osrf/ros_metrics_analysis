```python
# Extract the total pull for the "ROS" image by month
topline_ros = []
for f in all_files:
    print(f)
    df = pd.read_csv(f)
    #t = df[df["Repo"]=="ros"]["Pull by tag"].iloc[0]
    t = df[df["Repo"]=="ros"]["Unique IPs"].iloc[0]

    topline_ros.append(t)
```

```python
# For some reason the 10/2022 numbers are way off!
# Use 11/2022 as the start point 
print(topline_ros)
print("Total ROS unique IPs Nov 2022 to Oct 2023: {0}".format(np.array(topline_ros).sum()))
print("Total ROS unique IPs Nov 2022: {0}".format(topline_ros[0]))
print("Total ROS unique IPs Oct 2023: {0}".format(topline_ros[-1]))
print("Percent change in unique IPs: {0}".format(100.0*(topline_ros[-1]-topline_ros[0])/topline_ros[0]))
```
